```python
#Import libraries
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
import csv
```

```python
#Read weatherinformation JSON files
edinburgh = open('weatherUK/edinburgh.json', 'r')
liverpool = open('weatherUK/liverpool.json', 'r')
london = open('weatherUK/london.json', 'r')
notthingham = open('weatherUK/notthingham.json', 'r')
plymouth = open('weatherUK/plymouth.json', 'r')
text = open('days.txt', 'r')
```

```python
#Strip weeks+temperature from JSON files
def temperatureList(file):
    data = json.loads(file.read())
    temperature = []
    for x in data['data']:
        temperature.append(x['temperature'])
    return(temperature)
```

```python
#Create a DataFrame that contains average temperature measured over all weather stations per day
#This frame contains NaN values the weatherinformation JSON files are not complete
completedays = text.read().split()
daysFrame = pd.DataFrame({'week':completedays})
edin = pd.DataFrame({'edin':temperatureList(edinburgh)})
live = pd.DataFrame({'live':temperatureList(liverpool)})
lond = pd.DataFrame({'lond':temperatureList(london)})
nott = pd.DataFrame({'nott':temperatureList(notthingham)})
plym = pd.DataFrame({'plym':temperatureList(plymouth)})
data = daysFrame.join(edin).join(live).join(lond).join(nott).join(plym)
dayTemp = pd.DataFrame(data.set_index('week').transpose().mean())
```

```python
#Create a DataFrame that contains the average compound segment of all tracks in the top 200 per day
days = [f for f in listdir('dataGbDay') if isfile(join('dataGbDay', f))]
totalcomp = []
for file in days:
    with open('dataGbDay/' + file, newline='', encoding="utf-8") as csvfile:
        reader = csv.reader(csvfile)
        compound = []
        for index,row in enumerate(reader):
            try:
                compound.append(float(row[7]))
            except:
                compound.append(0)
        totalcomp.append(sum(compound)/len(compound))
totalcomp = daysFrame.join(pd.DataFrame(totalcomp))
```

```python
#Plot temperature and compound segment in one figure
result = pd.merge(dayTemp, totalcomp, on='week', suffixes=('Temp', 'Compound'))
result = result.drop('week', axis=1)

fig, temp = plt.subplots()

color = 'tab:red'
temp.set_xlabel('days')
temp.set_ylabel('Temperature')
temp.plot(result['0Temp'], color=color)
temp.tick_params(axis='y')

comp = temp.twinx()

color = 'tab:blue'
comp.set_ylabel('Compound Segment')
comp.plot(result['0Compound'], color=color)
comp.tick_params(axis='y')

fig.tight_layout()
plt.show()
```
