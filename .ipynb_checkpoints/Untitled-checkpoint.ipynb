### Q?: How do the characteristics of songs influence the sentiment
The dataframe based on indexes.csv (containing all unique songs that appeared in the top 200 in 2019) contains a lot of information provided by spotify. These include: key, danceability, energy, instrumentalness, liveness, loudness, speechiness, valence and tempo<br>
In this question we will try to find correlation between these characteristics of the songs and the compound segment<br><br>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_rows = 20
np.set_printoptions(precision = 4, suppress = True)
```

```python
this_weeks_top200_file = "indexes.csv"
df = pd.read_csv(this_weeks_top200_file)
```

```python
df.columns
```

The function chunker finds the minimum and maximum value of a certain column in the dataframe, based on these values we later split up the dataframe in groups.

```python
def chunker(column):
    mini = df[column].min()
    maxi = df[column].max()
    steps = (maxi-mini)/5
    chunks =[]
    for i in np.arange(mini, maxi, steps):
        chunks.append(i)
    chunks.append(maxi)
    return(chunks)
```

```python
chunker('energy')
```

As we can see in the example of the column 'energy', the function <i>chunker</i> returns [0.0541, 0.24188, 0.42966, 0.61744, 0.80522, 0.993] giving us the groups of:<br>
<=0.24188<br>
0.24188 - 0.42966<br>
0.42966 - 0.61744<br>
0.61744 - 0.80522<br>
.>= 0.993

The function <i>meanSentiment()</i> splits up the dataframe in chunks which are calculated in <i>chunker().</i><br>
It also calculates the mean of the given column (per chunk) and then plots the data to visualize and draw conclusions.

```python
def meanSentiment(column):
    chunks = chunker(column)
    very_low = df[df[column] <= chunks[1]]
    low = df.loc[(df[column] > chunks[1]) & (df[column] <= chunks[2])]
    medium = df.loc[(df[column] > chunks[2]) & (df[column] <= chunks[3])]
    high = df.loc[(df[column] > chunks[3]) & (df[column] <= chunks[4])]
    very_high = df.loc[(df[column] > chunks[4]) & (df[column] <= chunks[5])]

    very_low_seg = pd.to_numeric(very_low['compound'], errors='coerce').mean()
    low_seg = pd.to_numeric(low['compound'], errors='coerce').mean()
    medium_seg = pd.to_numeric(medium['compound'], errors='coerce').mean()
    high_seg = pd.to_numeric(high['compound'], errors='coerce').mean()
    very_high = pd.to_numeric(very_high['compound'], errors='coerce').mean()
    segments = [very_low_seg, low_seg, medium_seg, high_seg, very_high]
    
    langs = ['Very low', 'Low', 'Medium', 'High', 'Very high']
    plt.bar(langs,segments)

    plt.ylabel('Compound segment', fontsize=14)
    plt.xlabel(column, fontsize=12)
```

```python
meanSentiment('energy')
```

```python
meanSentiment('instrumentalness')
```

```python
meanSentiment('liveness')
```

```python
meanSentiment('loudness')
```

```python
meanSentiment('speechiness')
```

```python
meanSentiment('valence')
```

```python
meanSentiment('tempo')
```
